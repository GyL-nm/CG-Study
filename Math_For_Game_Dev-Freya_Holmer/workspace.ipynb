Beyond linear algebra, trigonometry, etc. more complex mathematical concepts tend to be rare (i.e. integrals, derivatives, etc.)

Numbers in the domain of computer graphics should always be thought of as not just numbers, but vectors - with a weight and a magnitude.

We can get the direction of the vector by checking its sign:

$sign(x) = direction [-1,1]$

The absolute value gives us the magnitude/length:

$abs(x) = magnitude$
